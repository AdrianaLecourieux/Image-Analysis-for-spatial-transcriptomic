# Venn Diagrams

In this notebook, we plot Venn Diagram of HVG between PDAC and healthy condition of PDAC dataset

The data are from **HVG_from_GraphST_clusters.ipynb**

The gene lists obtained via the diagrams are used in GOrilla in order to study gene ontology

# 1. Import packages and data

```python
!pip install matplotlib-venn
```

```python
import pandas as pd
from matplotlib_venn import venn2
import json
```

Import top 300 HGV in PDAC datasets:

```python
A1_300_HGV = pd.read_csv("/../../sbgenomics/project-files/HVG/A1_300_HGV.csv", index_col = 0)
B2_300_HGV = pd.read_csv("/../../sbgenomics/project-files/HVG/B2_300_HGV.csv", index_col = 0)
C3_300_HGV = pd.read_csv("/../../sbgenomics/project-files/HVG/C3_300_HGV.csv", index_col = 0)
D4_300_HGV = pd.read_csv("/../../sbgenomics/project-files/HVG/D4_300_HGV.csv", index_col = 0)
```

Get list of genes:

```python
A1_300_HGV_list = A1_300_HGV.index.values.tolist()
B2_300_HGV_list = B2_300_HGV.index.values.tolist()
C3_300_HGV_list = C3_300_HGV.index.values.tolist()
D4_300_HGV_list = D4_300_HGV.index.values.tolist()
```

# 2. Venn diagram of PDAC A1 and B2

```python
venn2([set(A1_300_HGV_list), set(B2_300_HGV_list)], ('PDAC A1', 'PDAC B2'))
```

# 3. Venn diagram of PanCN C3 and D4

```python
venn2([set(C3_300_HGV_list), set(D4_300_HGV_list)], ('PanCN C3', 'PanCN D4'), set_colors=('lightblue', 'yellow'))
```

# 4.Venn diagram of (all) PDAC and (all) PanCN (without duplicates)

Merged PDAC genes :

```python
A1_B2_merged_list = A1_300_HGV_list + B2_300_HGV_list
len(A1_B2_merged_list)
```

```python
A1_B2_merged_list_wo_duplicates = list(dict.fromkeys(A1_B2_merged_list))
len(A1_B2_merged_list_wo_duplicates)
```

Merged non-PDAC genes :

```python
C3_D4_merged_list = C3_300_HGV_list + D4_300_HGV_list
len(C3_D4_merged_list)
```

```python
C3_D4_merged_list_wo_duplicates = list(dict.fromkeys(C3_D4_merged_list))
len(C3_D4_merged_list_wo_duplicates)
```

Venn diagram between PDAC and PanCN genes:

```python
venn2([set(A1_B2_merged_list_wo_duplicates), set(C3_D4_merged_list_wo_duplicates)], ('PDAC', 'PanCN'), set_colors=('blue', 'green'))
```

```python
C3_D4_merged_list_wo_duplicates
```

```python
intersection_list = set(A1_B2_merged_list_wo_duplicates).intersection(C3_D4_merged_list_wo_duplicates)
len(intersection_list)
```

```python
PDAC_diff_list = set(A1_B2_merged_list_wo_duplicates).difference(C3_D4_merged_list_wo_duplicates)
```

```python
PDAC_diff_list
```

```python
PanCN_diff_list = set(C3_D4_merged_list_wo_duplicates).difference(A1_B2_merged_list_wo_duplicates)
```

```python
PanCN_diff_list
```

```python
len(PanCN_diff_list)
```

# 5. Venn diagram of (commun) PDAC and (commun) PanCN

```python
intersection_PDAC_list = set(A1_300_HGV_list).intersection(B2_300_HGV_list)
len(intersection_PDAC_list)
```

```python
intersection_PanCN_list = set(C3_300_HGV_list).intersection(D4_300_HGV_list)
len(intersection_PanCN_list)
```

```python
venn2([set(intersection_PDAC_list), set(intersection_PanCN_list)], ('PDAC', 'PanCN'), set_colors=('lightblue', 'lightgreen'))
```

```python
PDAC_diff_list_commun = set(intersection_PDAC_list).difference(intersection_PanCN_list)
len(PDAC_diff_list_commun)
```

```python
PDAC_diff_list_commun
```

```python
PDAC_diff_list_commun = list(PDAC_diff_list_commun)
```

```python
PDAC_diff_list = list(PDAC_diff_list)
```

# 6. Save list

```python
with open("PDAC_diff_list_commun.json", 'w') as f:
    # indent=2 is not needed but makes the file human-readable 
    # if the data is nested
    json.dump(PDAC_diff_list_commun, f, indent=2) 
```

```python
with open("PDAC_diff_list.json", 'w') as f:
    # indent=2 is not needed but makes the file human-readable 
    # if the data is nested
    json.dump(PDAC_diff_list, f, indent=2) 
```
